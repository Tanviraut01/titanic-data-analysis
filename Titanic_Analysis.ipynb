```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.shape        # rows, columns
df.columns      # column names
df.info()       # data types
df.describe()   # summary
```

```python
df.isnull().sum()     # check missing values
df.dropna(inplace=True)

df.duplicated().sum() # check duplicates
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt

df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Count")
plt.show()
```

```python
df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
df['Age'].plot(kind='hist')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
import sqlite3

conn = sqlite3.connect('titanic.db')
df.to_sql('titanic', conn, if_exists='replace', index=False)
```

```python
pd.read_sql("SELECT * FROM titanic LIMIT 5", conn)
```

```python
pd.read_sql("SELECT COUNT(*) FROM titanic", conn)
```

```python
pd.read_sql("SELECT Survived, COUNT(*) FROM titanic GROUP BY Survived", conn)
```

```python
pd.read_sql("""
SELECT Sex, AVG(Survived) as survival_rate
FROM titanic
GROUP BY Sex
""", conn)
```

```python
pd.read_sql("""
SELECT Pclass, AVG(Survived) as survival_rate
FROM titanic
GROUP BY Pclass
""", conn)
```
